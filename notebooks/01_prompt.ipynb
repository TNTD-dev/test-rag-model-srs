```python
# ==========================================
# 2. Äá»ŠNH NGHÄ¨A CÃC PROMPT Gá»C
# ==========================================

# Prompt táº¡o sinh SRS dá»±a trÃªn cáº¥u trÃºc máº«u cá»§a Innotech
SRS_SYSTEM_PROMPT = """
Báº¡n lÃ  má»™t Senior Business Analyst vÃ  Technical Architect vá»›i 10 nÄƒm kinh nghiá»‡m.
Nhiá»‡m vá»¥: Viáº¿t tÃ i liá»‡u Äáº·c táº£ yÃªu cáº§u pháº§n má»m (SRS) dá»±a trÃªn cáº¥u trÃºc chuáº©n chuyÃªn nghiá»‡p.

Cáº¥u trÃºc báº¯t buá»™c tuÃ¢n thá»§:
1. Giá»›i thiá»‡u: Má»¥c Ä‘Ã­ch, Pháº¡m vi (Table), Thuáº­t ngá»¯.
2. MÃ´ táº£ tá»•ng quan: SÆ¡ Ä‘á»“ kiáº¿n trÃºc (Mermaid), NhÃ³m ngÆ°á»i dÃ¹ng (Table).
3. Khung Workflow Framework (TÃ¡i sá»­ dá»¥ng): 
   - Äá»‹nh nghÄ©a cÃ¡c Engine (Approval, Notification, Automation, Scheduling).
   - CÃ¡c Workflow Patterns (Sequential, Parallel, Matrix...).
   - Database Schema máº«u cho Framework.
4. CÃ¡c Module chá»©c nÄƒng: 
   - Má»—i module pháº£i cÃ³: Business Flow (Mermaid), Activity Diagram, DB Schema, vÃ  Business Rules (Ä‘Ã¡nh mÃ£ XXX-BR-01).
5. YÃªu cáº§u phi chá»©c nÄƒng: Hiá»‡u nÄƒng, Báº£o máº­t, Äá»™ tin cáº­y.

VÄƒn phong: ChuyÃªn nghiá»‡p, dÃ¹ng Ä‘Ãºng thuáº­t ngá»¯ chuyÃªn ngÃ nh IT/BA. TrÃ¡nh bá»‹a Ä‘áº·t (Anti-Hallucination).
"""

# Prompt Ä‘Ã¡nh giÃ¡ dá»±a trÃªn tÃ i liá»‡u LLM-as-judge.md
EVALUATION_PROMPT_TEMPLATE = """
Báº¡n lÃ  má»™t "GiÃ¡m kháº£o AI" chuyÃªn nghiá»‡p. HÃ£y Ä‘Ã¡nh giÃ¡ báº£n SRS sau Ä‘Ã¢y dá»±a trÃªn cÃ¡c tiÃªu chÃ­ nghiÃªm ngáº·t:

### Bá»™ tiÃªu chÃ­ cháº¥m Ä‘iá»ƒm (Scale 1-10):
1. TÃ­nh Ä‘áº§y Ä‘á»§ (Completeness - Trá»ng sá»‘ 30%): CÃ³ Ä‘á»§ cÃ¡c má»¥c Giá»›i thiá»‡u, Workflow, Module, DB Schema, Phi chá»©c nÄƒng khÃ´ng?
2. TÃ­nh nháº¥t quÃ¡n (Consistency - Trá»ng sá»‘ 30%): Logic giá»¯a cÃ¡c pháº§n cÃ³ mÃ¢u thuáº«n khÃ´ng? (VÃ­ dá»¥: Quy trÃ¬nh mÃ´ táº£ cÃ³ khá»›p vá»›i DB Schema khÃ´ng?)
3. Äá»™ chÃ­nh xÃ¡c nghiá»‡p vá»¥ (Accuracy - Trá»ng sá»‘ 20%): Sá»­ dá»¥ng Ä‘Ãºng thuáº­t ngá»¯ BA/IT khÃ´ng? Quy trÃ¬nh cÃ³ thá»±c táº¿ khÃ´ng?
4. Äá»‹nh dáº¡ng & VÄƒn phong (Format & Tone - Trá»ng sá»‘ 20%): CÃ³ dÃ¹ng Markdown chuáº©n vÃ  Mermaid khÃ´ng? CÃ³ bá»‹ "bá»‹a Ä‘áº·t" thÃ´ng tin khÃ´ng?

### YÃªu cáº§u Ä‘áº§u ra:
- Äiá»ƒm chi tiáº¿t cho tá»«ng tiÃªu chÃ­.
- Tá»•ng Ä‘iá»ƒm trá»ng sá»‘.
- PhÃ¢n loáº¡i: 9-10 (Xuáº¥t sáº¯c), 7-8 (Tá»‘t), 5-6 (Trung bÃ¬nh), <5 (KÃ©m).
- Nháº­n xÃ©t cá»¥ thá»ƒ nhá»¯ng Ä‘iá»ƒm cáº§n cáº£i thiá»‡n.

---
Báº¢N SRS Cáº¦N ÄÃNH GIÃ:
{srs_content}
"""
```

```python
# ==========================================
# 3. CÃC HÃ€M Xá»¬ LÃ CHÃNH
# ==========================================

def generate_srs(project_description):
    print("...Äang táº¡o sinh SRS vá»›i mÃ´ hÃ¬nh Qwen3-Coder...")
    response = client.chat.completions.create(
        model="Qwen/Qwen3-Coder-30B-A3B-Instruct:nebius", 
        messages=[
            {"role": "system", "content": SRS_SYSTEM_PROMPT},
            {"role": "user", "content": f"HÃ£y viáº¿t SRS cho dá»± Ã¡n sau: {project_description}"}
        ],
        temperature=0.2,
    )
    return response.choices[0].message.content

def evaluate_srs(srs_content):
    print("...Gemini Ä‘ang tiáº¿n hÃ nh cháº¥m Ä‘iá»ƒm...")
    full_eval_prompt = EVALUATION_PROMPT_TEMPLATE.format(srs_content=srs_content)
    response = gemini_model.generate_content(full_eval_prompt)
    return response.text
```

```python
project_input = input("Nháº­p mÃ´ táº£ dá»± Ã¡n cá»§a báº¡n (vÃ­ dá»¥: Há»‡ thá»‘ng quáº£n lÃ½ kho thÃ´ng minh): ")
    
if not project_input.strip():
    print("Vui lÃ²ng nháº­p mÃ´ táº£ dá»± Ã¡n!")
    
    
# BÆ°á»›c 1: Táº¡o sinh
generated_srs = generate_srs(project_input)
    
display(Markdown("## ğŸ“„ Báº¢N SRS ÄÆ¯á»¢C Táº O SINH"))
display(Markdown(generated_srs))
    
print("-" * 50)
```

```python
evaluation_result = evaluate_srs(generated_srs)
    
display(Markdown("## LLM-as-Judge"))
display(Markdown(evaluation_result))
```

```python
import pandas as pd
import time

final_results = []

print(f"Báº¯t Ä‘áº§u quy trÃ¬nh xá»­ lÃ½ {len(project_inputs)} máº«u...")

for i, input_text in enumerate(project_inputs):
    print(f"\n--- Äang xá»­ lÃ½ Máº«u {i+1}/10 ---")
    try:
        # 1. Táº¡o sinh SRS báº±ng Qwen
        srs_content = generate_srs(input_text)
        
        # 2. ÄÃ¡nh giÃ¡ báº±ng Gemini
        evaluation = evaluate_srs(srs_content)
        
        # 3. LÆ°u káº¿t quáº£ vÃ o danh sÃ¡ch
        final_results.append({
            "Input": input_text,
            "Output_SRS": srs_content,
            "Evaluation_Judge": evaluation,
            "Status": "Success"
        })
        
        # Hiá»ƒn thá»‹ tiáº¿n Ä‘á»™ ngáº¯n gá»n
        print(f"HoÃ n thÃ nh máº«u {i+1}. Äang nghá»‰ 5 giÃ¢y Ä‘á»ƒ trÃ¡nh Rate Limit...")
        time.sleep(5) # Nghá»‰ Ä‘á»ƒ Ä‘áº£m báº£o khÃ´ng vi pháº¡m giá»›i háº¡n gá»i API
        
    except Exception as e:
        print(f"Lá»—i á»Ÿ máº«u {i+1}: {str(e)}")
        final_results.append({
            "Input": input_text,
            "Output_SRS": "N/A",
            "Evaluation_Judge": f"Error: {str(e)}",
            "Status": "Failed"
        })

# 4. Chuyá»ƒn Ä‘á»•i sang DataFrame vÃ  xuáº¥t file CSV
df_results = pd.DataFrame(final_results)
csv_filename = "srs_prompt_evaluation_report.csv"
df_results.to_csv(csv_filename, index=False, encoding='utf-8-sig')

print(f"\n========================================")
print(f"Táº¤T Cáº¢ HOÃ€N Táº¤T! ÄÃ£ lÆ°u káº¿t quáº£ táº¡i: {csv_filename}")
```

```python
import pandas as pd
import re
import os

def extract_score_and_group(text):
    if not isinstance(text, str):
        return pd.Series([None, None], index=['Score', 'Group'])

    # 1. TrÃ­ch xuáº¥t Äiá»ƒm (Score)
    # TÃ¬m máº«u: **8.2 / 10** hoáº·c **8.2/10**
    score_match = re.search(r'\*\*\s*([\d\.]+)\s*/\s*10\s*\*\*', text)
    score = float(score_match.group(1)) if score_match else None

    # 2. TrÃ­ch xuáº¥t PhÃ¢n loáº¡i (Group)
    group = None
    # TÃ¬m vá»‹ trÃ­ cá»§a má»¥c "PhÃ¢n loáº¡i" (khÃ´ng phÃ¢n biá»‡t hoa thÆ°á»ng)
    classification_match = re.search(r'(?:III\.|###)\s*(?:PhÃ¢n loáº¡i|PHÃ‚N LOáº I)', text, re.IGNORECASE)
    
    if classification_match:
        # Láº¥y pháº§n vÄƒn báº£n tá»« sau chá»¯ "PhÃ¢n loáº¡i" Ä‘áº¿n háº¿t
        post_classification_text = text[classification_match.end():]
        
        # CÃ¡ch 1: TÃ¬m cÃ¡c tá»« khÃ³a xáº¿p loáº¡i chuáº©n (Æ°u tiÃªn Ä‘á»™ chÃ­nh xÃ¡c)
        keywords = ["XUáº¤T Sáº®C", "Tá»T", "KHÃ", "TRUNG BÃŒNH", "KÃ‰M"]
        found_keywords = [kw for kw in keywords if kw in post_classification_text.upper()]
        
        if found_keywords:
            group = found_keywords[0]
        else:
            # CÃ¡ch 2: Láº¥y ná»™i dung trong cáº·p dáº¥u ** ** ngay sau Ä‘Ã³
            bold_match = re.search(r'\*\*(.*?)\*\*', post_classification_text)
            if bold_match:
                group = bold_match.group(1).strip()
    
    return pd.Series([score, group], index=['Score', 'Group'])

# ÄÆ°á»ng dáº«n file CSV
csv_file_path = 'srs_prompt_evaluation_report.csv'

# Kiá»ƒm tra file cÃ³ tá»“n táº¡i khÃ´ng
if os.path.exists(csv_file_path):
    print(f"Äang Ä‘á»c file: {csv_file_path}")
    df = pd.read_csv(csv_file_path)
    
    # Kiá»ƒm tra xem cá»™t Evaluation_Judge cÃ³ tá»“n táº¡i khÃ´ng
    if 'Evaluation_Judge' in df.columns:
        print("Äang trÃ­ch xuáº¥t dá»¯ liá»‡u...")
        df[['Score', 'Group']] = df['Evaluation_Judge'].apply(extract_score_and_group)
        
        # LÆ°u Ä‘Ã¨ lÃªn file cÅ©
        df.to_csv(csv_file_path, index=False, encoding='utf-8-sig')
        print(f"ThÃ nh cÃ´ng! File Ä‘Ã£ Ä‘Æ°á»£c cáº­p nháº­t: {csv_file_path}")
        print("\n5 dÃ²ng Ä‘áº§u tiÃªn:")
        print(df[['Score', 'Group']].head())
    else:
        print("Lá»—i: KhÃ´ng tÃ¬m tháº¥y cá»™t 'Evaluation_Judge' trong file CSV.")
else:
    print(f"Lá»—i: KhÃ´ng tÃ¬m tháº¥y file {csv_file_path}")
```
